```python
missing_values = ['nan','NAN','na','nil','NaN','n/a']
df1 = df1.replace(missing_values,np.NaN)
df1["Murder"].unique()
```

```python
df1['Murder'] = df1['Murder'].fillna(df1.groupby('STATE/UT')['Murder'].transform('median'))
df1['Rape'] = df1['Rape'].fillna(df1.groupby('STATE/UT')['Rape'].transform('mean'))
df1['Kidnapping and Abduction'] = df1['Kidnapping and Abduction'].fillna(df1.groupby('STATE/UT')['Kidnapping and Abduction'].transform('median'))
df1['Foeticide'] = df1['Foeticide'].fillna(df1.groupby('STATE/UT')['Foeticide'].transform('median'))
df1['Abetment of suicide'] = df1['Abetment of suicide'].fillna(df1.groupby('STATE/UT')['Abetment of suicide'].transform('median'))
df1['Exposure and abandonment'] = df1['Exposure and abandonment'].fillna(df1.groupby('STATE/UT')['Exposure and abandonment'].transform('median'))
df1['Procuration of minor girls'] = df1['Procuration of minor girls'].fillna(df1.groupby('STATE/UT')['Procuration of minor girls'].transform('median'))
df1['Buying of girls for prostitution'] = df1['Buying of girls for prostitution'].fillna(df1.groupby('STATE/UT')['Buying of girls for prostitution'].transform('median'))
df1['Selling of girls for prostitution'] = df1['Selling of girls for prostitution'].fillna(df1.groupby('STATE/UT')['Selling of girls for prostitution'].transform('median'))
df1['Prohibition of child marriage act'] = df1['Prohibition of child marriage act'].fillna(df1.groupby('STATE/UT')['Prohibition of child marriage act'].transform('median'))
df1['Other Crimes'] = df1['Other Crimes'].fillna(df1.groupby('STATE/UT')['Other Crimes'].transform('median'))
```

```python
unique_murders = df1["Murder"].unique()
print(unique_murders)
```

```python
data_types_dict = {'Infanticid':float,'Other murder':float,'Rape':float,'Kidnapping and Abduction':float,'Foeticide':float, 'Abetment of suicide':float, 'Exposure and abandonment':float, 'Procuration of minor girls': float, 'Buying of girls for prostitution':float, 'Selling of girls for prostitution':float, 'Prohibition of child marriage act':float, 'Other Crimes':float} 
df2 = df2.astype(data_types_dict)
data_types_dict1 = {'Total':float}
df1 = df1.astype(data_types_dict1)
df2.info()
```

```python
df2 = df2.drop(['Other murder','Infanticid'],axis=1)
df1 = df1.drop(['Murder'],axis=1)
```

```python
df_final = pd.concat([df1,df2],axis=0)
df_final
```

```python
Rape = df_final['Rape'].sum()
KA = df_final['Kidnapping and Abduction'].sum()
F = df_final['Foeticide'].sum()
AS = df_final['Abetment of suicide'].sum()
EA = df_final['Exposure and abandonment'].sum()
PMG = df_final['Procuration of minor girls'].sum()
BGP = df_final['Buying of girls for prostitution'].sum()
SGP = df_final['Selling of girls for prostitution'].sum()
PFMA = df_final['Prohibition of child marriage act'].sum()
OC = df_final['Other Crimes'].sum()

import plotly.graph_objects as go

grp = ['Rape','Kidnapping and Abduction','Foeticide','Abetment of suicide','Exposure and abandonment','Buying of girls for prostitution','Selling of girls for prostitution','Prohibition of child marriage act','Other Crimes']
group_vals = [Rape,KA,F,AS,EA,PMG,BGP,SGP,PFMA,OC]

fig = go.Figure(data=[go.Pie(labels=grp, values=group_vals,sort=True,
                            marker=dict(colors=px.colors.qualitative.G10),textfont_size=12)])

fig.show()
```
